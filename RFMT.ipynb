```python
# Create snapshot date
snapshot_date = df4['ts_order_purchase'].max() + timedelta(days=1)
print(snapshot_date)
```

```python
# Grouping by CustomerID
data_process = df4.groupby(['customer_unique_id']).agg({
    'ts_order_purchase': lambda x: (snapshot_date - x.max()).days})
```

```python
data_process = data_process.reset_index()
```

```python
data_process1 = df4[['order_count','tot_spending', 'customer_unique_id', "tenure",'tot_ship/tot_spend','tot_shipping_spending']]
```

```python
data_process = pd.merge(data_process1, data_process, on = 'customer_unique_id', how='right')
```

```python
data_process.isna().sum()
```

```python
data_process = data_process.drop_duplicates(['customer_unique_id'],keep= 'last')
```

```python
#Relative frequency so that we avoid frequency bias (link in NTT doc)
data_process['RelRecency'] = data_process['ts_order_purchase'] / (data_process['tenure'])
data_process['RelFrequency'] = data_process['order_count'] / (data_process['tenure'])
data_process['RelMonetary'] = data_process['tot_spending'] / (data_process['tenure'])
```

```python
# Rename the columns 
data_process.rename(columns={'ts_order_purchase': 'Recency',
                         'order_count': 'Frequency',
                         'tot_spending': 'MonetaryValue'}, inplace=True)
```

```python
data_process = data_process.set_index('customer_unique_id')
```

```python
# --Calculate R and F groups--
# Create labels for Recency 
r_labels = range(4, 0, -1)
# Assign these labels to 4 equal percentile groups 
r_groups = pd.qcut(data_process['RelRecency'], q=4, labels=r_labels)
# Create new columns R 
data_process = data_process.assign(R = r_groups.values)
data_process.head()
```

```python
f_labels = range(1, 7)
# Assign these labels to 4 equal percentile groups 
f_groups = pd.cut(data_process['RelFrequency'], 6, labels=f_labels, duplicates='drop')
# Create new columns F 
data_process = data_process.assign(F = f_groups.values)
```

```python
data_process[data_process['F'] == 6]
```

```python
# Create labels for MonetaryValue
m_labels = range(1, 6)
# Assign these labels to three equal percentile groups 
m_groups = pd.qcut(data_process['RelMonetary'], 5, labels=m_labels)
# Create new column M
data_process = data_process.assign(M = m_groups.values)
```

```python
data_process['R']=data_process.R.astype(np.int64)
data_process['F']=data_process.F.astype(np.int64)
data_process['M']=data_process.M.astype(np.int64)
```

```python
# Concat RFM quartile values to create RFM Segments
def join_rfm(x): return str(x['R']) + str(x['F']) + str(x['M'])
data_process['RFM_Segment_Concat'] = data_process.apply(join_rfm, axis=1)
rfm = data_process
rfm.head()
```

```python
# Count num of unique segments
rfm_count_unique = rfm.groupby('RFM_Segment_Concat')['RFM_Segment_Concat'].nunique()
print(rfm_count_unique.sum())
```

```python
# Calculate RFM_Score
rfm['RFM_Score'] = rfm[['R','F','M']].sum(axis=1)
print(rfm['RFM_Score'].head())
```

```python
# Define rfm_level function
def rfm_level(df):
    if df['RFM_Score'] >= 9:
        return 'Can\'t Loose Them'
    elif ((df['RFM_Score'] >= 8) and (df['RFM_Score'] < 9)):
        return 'Champions'
    elif ((df['RFM_Score'] >= 7) and (df['RFM_Score'] < 8)):
        return 'Loyal'
    elif ((df['RFM_Score'] >= 6) and (df['RFM_Score'] < 7)):
        return 'Potential'
    elif ((df['RFM_Score'] >= 5) and (df['RFM_Score'] < 6)):
        return 'Promising'
    elif ((df['RFM_Score'] >= 4) and (df['RFM_Score'] < 5)):
        return 'Needs Attention'
    else:
        return 'Require Activation'
# Create a new variable RFM_Level
rfm['RFM_Level'] = rfm.apply(rfm_level, axis=1)
# Print the header with top 5 rows to the console
rfm.head()
```
